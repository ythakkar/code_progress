```python
##Libraries
import json
import numpy as np
```

```python
## Import data
with open('face-rating-data.json') as desc_file:
    descr = json.load(desc_file)
```

```python
##Check data
print(len(descr))
print(descr[0:2])

## A list of dictionaries. Each dictionary describes one of the faces. One face could be described 
## by various participants. So the primary key here is face ID. Join all the description of each 
## face_id(as a string) in a list. Then add all the lists of descriptions together in big list
```

```python
##distinct faceIDS
faceids = []
for trial in descr:
    faceids.append(trial['faceID'])

##check length    
print(len(faceids))

## distinct face ids
dist_faceid = list(set(faceids))

## check length
print(len(dist_faceid))
```

```python
## add descriptions for each face id:
faceid_des = []
for f in dist_faceid:
    all_descr = ''
    for d in descr:
        if f == d ['faceID']:
            all_info = d['responses']
            my_descr = all_info['description'] + " Age is " + str(all_info['age'] + '. ')
            all_descr += my_descr
    faceid_des.append(all_descr)
    
        
        
print(dist_faceid[0:2])   
print(len(faceid_des))
print(faceid_des[0])
```

```python
descr_by_id = [[i, j] for i, j in zip(dist_faceid,faceid_des)]
```

```python
## check descr_by_id
print(descr_by_id)
```

```python
np.savetxt('face_id_descr.csv', descr_by_id, fmt='%s', delimiter=',', 
           header = "faceid, description", comments = "")
```
